```python
import networkx as nxs
from networkx.generators.community import LFR_benchmark_graph

from seaborn import color_palette, set_style, palplot
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
G = nx.relaxed_caveman_graph(3, 12, 0.1, seed=42)
G.remove_edges_from(nx.selfloop_edges(G))

ax, fig = plt.subplots(figsize=(20,15))
colors = [(0, 0,0,.1) for i in range(300)]
node_color = '#597769'
nx.draw(G, node_size = 480, with_labels=False, node_color=node_color, edge_color=colors, edge_cmap=plt.cm.Blues)

plt.plot()

plt.savefig('network-caveman-480.png', bbox_inches='tight', transparent=True)
```

```python
G = nx.planted_partition_graph(5, 80, 0.3, 0.01, seed=42)
```

```python
deg_centrality = nx.degree_centrality(G)
ax, fig = plt.subplots(figsize=(20,15))
colors = [(0,0,0,.02) for i in range(400)]

nx.draw(G, node_size = 70, with_labels=False, node_color=list(deg_centrality.values()),
        edge_color=colors, cmap=cm.viridis)


plt.plot()
plt.savefig('network-ppg.png', bbox_inches='tight', transparent=True)
```
